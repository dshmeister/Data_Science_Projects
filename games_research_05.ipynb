## Шаг 5. Проверьте гипотезы

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

Здесь необходимо построить гипотезы и либо доказать их, либо опровергнуть

- **H0** - средние пользовательские рейтинги XOne и PC одинаковые
- **H1** - средние пользовательские рейтинги XOne и PC разные

Уровень значимости 0.01

```python
import scipy.stats as st

x_box = data_act.loc[data_act['platform']=='XOne']
pc = data_act.loc[data_act['platform']=='PC']

k = 0.01
result = st.ttest_ind(x_box['user_score'].dropna(), pc['user_score'].dropna(), equal_var=True)
if result[1] <= k:
    print('Нулевая гипотеза отвергается. p-value:', result)
else:
    print('Нулевая гипотеза не отвергается. p-value:', result)
    
```

Нулевая гипотеза не отвергается при нашем уровне значимости, а значит, что можно утверждать, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

- **H0** - средние пользовательские рейтинги `Action` и `Sports` одинаковые
- **H1** - средние пользовательские рейтинги `Action` и `Sports` разные

Уровень статистической значимости оставим равным 0.05
Для проверки будем использовать ttest_ind для сравнения двух выборок.

```python
action = data_act.loc[data_act['genre'] == 'Action']
sports = data_act.loc[data_act['genre'] == 'Sports']

k = 0.05
result = st.ttest_ind(action['user_score'].dropna(), sports['user_score'].dropna(), equal_var = True)
if result[1] <= k:
    print('Нулевая гипотеза отвергается, значение p-value:', result[1])
else:
    print('Нулевая гипотеза не отвергается, значение p-value:', result[1])
```

Мы получили значение p-value равное 0.23, при нашем уровнее статистической значимости 0.05, что говорит о том, что нулевая гипотеза отвергается (средние пользовательские рейтинги `Action` и `Sports` одинаковые)

Краткий вывод:

В этом разделе необходимо было проверить следующие гипотезы:

- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- Средние пользовательские рейтинги жанров Action и Sports разные.

После получения значений для p_value, после проведения двустороннего ttest_ind теста для двух выборок мы получили следующие значения: 0,55 и 4,8, что говорит о том, что средние пользовательствие рейтинги платформ XOne, PC - одинаковые, и что средние оценки для Action и Sports разные.
